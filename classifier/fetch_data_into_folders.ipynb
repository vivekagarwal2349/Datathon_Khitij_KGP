```
import pandas as pd
from PIL import Image
import requests
from io import BytesIO
from google.colab import files


```

```
df = pd.read_csv('/content/drive/Team Drives/DataScienceFinalProject/final_data.csv')
```

```
path = "/content/drive/My Drive/car_types"

a = 1;
for index, row in df.iterrows():
    if (row['image-type'] == 'original'):
      response = requests.get("https://" + row['src'])
      img = Image.open(BytesIO(response.content))
      a = a + 1
      img = img.transpose(Image.FLIP_LEFT_RIGHT)
      img = img.resize((128,128), Image.ANTIALIAS)
      if (row['type'] == 'suv'):
        img.save(path + '/suv/' + str(a) + '.png')
      elif (row['type'] == 'sedan'):
        img.save(path + '/sedan/' + str(a) + '.png')
      elif (row['type'] == 'convertible' or row['type'] == 'coupe'):
        img.save(path + '/coupe/' + str(a) + '.png')
      elif (row['type'] == 'truck'):
        img.save(path + '/truck/' + str(a) + '.png') 
        
        
```
